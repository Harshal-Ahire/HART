```python
df = df.drop_duplicates(inplace = True)
```

```python
#how chol impact heart disease
#chol vs target

bins = [0, 200, 240, 300, np.inf]
labels = ['Low (<200)', 'Borderline (200-240)', 'High (240-300)', 'Very High (>300)']

# Create binned cholesterol column
df['chol_bin'] = pd.cut(df['chol'], bins=bins, labels=labels)

# Calculate the rate of heart disease for each bin
chol_heart_rate = df.groupby('chol_bin')['target'].mean().reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='chol_bin', y='target', data=chol_heart_rate, ci=True)
plt.title('Heart Disease Rate by Cholesterol Level')
plt.xlabel('Cholesterol Level (Binned)')
plt.ylabel('Heart Disease Rate (Proportion)')
plt.show()


```

```python
# Create bins for thalach
thalach_bins = [60, 100, 120, 140, 160, np.inf]
thalach_labels = ['50-100', '100-120', '120-140', '140-160', '160+']

df['thalach_bin'] = pd.cut(df['thalach'], bins=thalach_bins, labels=thalach_labels)

# Calculate the mean target (heart disease rate) for each thalach bin
thalach_heart_rate = df.groupby('thalach_bin')['target'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='thalach_bin', y='target', data=thalach_heart_rate, ci=True)
plt.title('Heart Disease Rate by Maximum Heart Rate Achieved')
plt.xlabel('Maximum Heart Rate Achieved (Binned)')
plt.ylabel('Heart Disease Rate (Proportion)')
plt.show()

```

```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Create a count plot to show the distribution of CA values for each Target category
plt.figure(figsize=(8, 6))
sns.countplot(x='ca', hue='target', data=df, palette='viridis')

# Add titles and labels
plt.title('Distribution of Number of Blocked Arteries (CA) by Heart Disease Target', fontsize=14)
plt.xlabel('Number of Major Blocked Arteries (CA)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Target', labels=['No Heart Disease (0)', 'Heart Disease (1)'])

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='age', hue='ca', data=df, palette='viridis')

# Add titles and labels
plt.title('Number of Blocked Arteries (CA) by Age Group', fontsize=14)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Number of Blocked Arteries (CA)')

# Show the plot
plt.show()
```
